# Demo

## Download Data

We need to specify the tickers of the assets we want to download data from, a start and end date, and the interval.

```python
tickers = ['SPY', 'AAPL', 'TSLA', 'KO']
data = aa.download_data(tickers, '2023-01-01', '2024-01-01', '1d')
data
```

## Get Data

We can get the personalized returns for each ticker.

```python
data_ret = aa.pct_returns(data,
                          values={'Daily':    ('Close', 'Close', 1),
                                  'Intraday': ('Close', 'Open' , 0),
                                  'HL_diff':  ('High' , 'Low'  , 0)})
data_ret.head(10)
```

If we have daily data (interval='1d') we can use this function to get various returns of it.

```python
data_ret_d = aa.pct_returns_daily(data)
data_ret_d.head(10)
```

Get some insights about the returns with this function.

```python
data_ins = aa.insights(data_ret)
data_ins
```

Print the top and bottom results of the insights.

```python
aa.top_results(data_ins, top=2, bot=1, round_val=3)
```

Normalize the prices and volume for future comparison.

```python
data_norm = aa.normalize_values(data)
data_norm
```

## Drop Columns, Get Date Values

We can drop a column of the second level that we know we will not use, like 'Adj Close'

```python
data_drop = aa.drop_columns(data, 'Adj Close')
data_drop.head(3)
```

This function gets a value like the day of the week or the quarter from the date index.

```python
data_date = aa.from_datetime(data, 'weekday', add_col=True)
data_date
```

## Plots

Plot the price and moving averages.

```python
aa.plot_price(data, 'SPY')
```

Plot the volume and a moving average.

```python
aa.plot_volume(data, 'SPY')
```

Combine the two plots above, price and volume in a single visualization.

```python
aa.plot_price_vol(data, 'SPY')
```

Compare the normalized movements of the assets.

```python
aa.plot_norm_moves(data_norm)
```

Or the their normalized volumes.

```python
aa.plot_norm_vol(data_norm)
```

We can visualize the distribution of the returns and compare with others.

```python
aa.plot_dist_ret(data_ret)
```

## Calculations

Calculate the Sharpe Ratio.

```python
aa.sharpe_ratio(data, 'SPY')
```

See the correlations between assets.

```python
aa.correlations(data)
```

Calculate the Relative Strength Index (RSI)

```python
aa.calc_rsi(data, 'SPY', add_col=True)
```

See its historical volatility.

```python
aa.calc_historical_volatility(data, 'SPY', add_col=True)
```

Plot the Bollinger Bands

```python
aa.calc_bollinger_bands(data, 'SPY', add_col=True)
```

Calculate the Moving Average Convergence Divergence, signal and its histogram.

```python
aa.calc_macd(data, 'SPY', add_col=True)
```

Get the mean monthly returns

```python
aa.monthly_returns(data, 'SPY')
```

With hourly data (interval='1h' or '60m') we can get the mean hourly returns.

```python
data_h = aa.download_data(tickers, '2022-07-01', '2024-05-01', '60m')

hourly_ret = aa.hourly_returns(data_h, 'SPY')
hourly_ret
```
